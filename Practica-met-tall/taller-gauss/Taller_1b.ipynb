### Eliminación Gaussiana 

  Existe dos fases donde se realizan las operaciones:

$i)$ Eliminación hacia adelante

Para cada operaciones de filas se ejecutan restas del tipo:

  $ A_{i,k} = A_{j,k} - m * A_{i,k} $

Lo que implica

- 1 Multiplicación 

- 1 Resta

Y sabemos que el numero total de operaciones crece aproximadamente como:

$O(n^3) $

$ii)$ Sustitución hacia atrás

Aquí en cambio sabemos que se ejecuta 

- sumas acumulativas

- restas finales

-divisiones

Y este costo creces como:

$O(n^2)$

### Eliminación Gauss-Jordan

Este metodo elimina tanto por debajo como por encima del pivote, por lo que:

- Existe la duplicación de operaciones con respecto a Gauss

- El número de sumas y restas es mayor

- Compare la complejidad computacional del método de eliminación gaussiana vs Gauss-Jordan.


```python
import numpy as np
import matplotlib.pyplot as plt

def gaussjordan(n):
    return (n ** 3) / 2 + n**2 - n / 2

def gaussia(n):
    return (n ** 3) / 3 + n**2 - n / 3

# Crear valores de x
x = np.linspace(0, 10, 100)

# Calcular valores de y
y1 = gaussjordan(x)
y2 = gaussia(x)

# Graficar
plt.plot(x, y1, label='Gauss-Jordan')
plt.plot(x, y2, label='Gaussia')
plt.xlabel('n')
plt.ylabel('Resultado')
plt.legend()
plt.grid()
plt.show()

```

Para el análisis de la complejidad utilizamos las expresiones ya implementadas en el código de arriba

- Método Gaussiana

Este mantiene una complejidad de $O(\frac{1}{3}n^3)$

- Método Gauss-Jordan

A cambio este método tiene una complejidad del tipo $O(\frac{1}{2}n^3)$

- En conclusión:

 * Gauss-Jordan es computacionalmente más cosotoso

 * Gaussiana es más eficiente, especialmente para sistemas mas grandes

Esto lo confirmamos visulamnete con la gráfica que se genera con las expresiones establecidas
